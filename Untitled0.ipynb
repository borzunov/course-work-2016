```python
from collections import defaultdict


with open('compilation.txt') as f:
    lines = f.readlines()

lines = filter(lambda x: 'Warning' not in x and 'length' not in x and 'Time' not in x, lines)
words = set()
stat_by_len = defaultdict(lambda: 0)
stat2_by_len = defaultdict(lambda: 0)
for line in lines:
    w, l, z = line.split()
    l = int(l)
    z = int(z)
    
    words.add((w, l, z))
    length = len(w)
    stat_by_len[length] += 1
    if l - z >= 2:
        stat2_by_len[length] += 1
```

```python
import matplotlib.pyplot as plt
import math

xs, ys = zip(*sorted(stat_by_len.items()))
plt.plot(xs, ys)

xs, ys = zip(*sorted(stat2_by_len.items()))
plt.plot(xs, ys)

# k = math.exp(2 - 13 * 3.05)
# xs, ys = zip(*[(i, k * i ** 13) for i in range(start_i, 35)])
# plt.plot(xs, ys, 'r')

plt.show()

print(list(sorted(stat_by_len.items())))
print(list(sorted(stat2_by_len.items())))
```

```python
max_len = max(stat_by_len.keys())
i_range = range(19, max_len + 1)

xs, ys = zip(*[(math.log(i), math.log(2 ** i)) for i in i_range])
plt.plot(xs, ys, 'or-')

xs, ys = zip(*[(math.log(i), math.log(stat_by_len[i])) for i in i_range if stat_by_len[i] > 0])
plt.plot(xs, ys, 'og-')

xs, ys = zip(*[(math.log(i), math.log(stat2_by_len[i])) for i in i_range if stat2_by_len[i] > 0])
plt.plot(xs, ys, 'ob-')

plt.show()

xs, ys = zip(*[(i, stat_by_len[i] / 2 ** i) for i in range(22, max_len + 1)])
plt.plot(xs, ys, 'o-')
plt.show()
```
