```python
from collections import defaultdict

with open('result.txt') as f:
    lines = f.readlines()

lines = filter(lambda x: 'length' not in x and '...' not in x, lines)
words = set()
stats = defaultdict(lambda: defaultdict(lambda: 0))
for line in lines:
    w, l, z = line.split()
    l = int(l)
    z = int(z)
    
    words.add((w, l, z))
    stats[l - z][len(w)] += 1
words = list(sorted(words, key=lambda info: len(info[0])))
```

```python
for diff in sorted(stats):
    print('L(S) - Z(S) =', diff)
    print(sorted(stats[diff].items()))
    print()

for n in range(14, 35):
    print(r'{} & {} & {} \\'.format(n, stats[1][n], stats[2][n]))
```

```python
import matplotlib.pyplot as plt

legend = ['L(S) - Z(S) = {}'.format(diff) for diff in sorted(stats)]

plt.figure(figsize=(4, 4))
plt.xlabel('Длина слова')
plt.ylabel('Кол-во слов со свойством')
lines = []
for diff in sorted(stats):
    xs, ys = zip(*sorted(stats[diff].items()))
    lines.append(plt.plot(xs, ys)[0])
plt.legend(lines, legend, loc='upper left')
plt.show()

plt.figure(figsize=(4, 4))
plt.xlabel('Длина слова')
plt.ylabel('Доля слов со свойством')
lines = []
for diff in sorted(stats):
    xs, ys = zip(*[(n, count / 2 ** n) for n, count in sorted(stats[diff].items()) if n >= 22])
    lines.append(plt.plot(xs, ys)[0])
plt.legend(lines, legend)
plt.show()
```
